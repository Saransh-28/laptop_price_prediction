```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split as tts
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('laptop_price.csv', encoding='latin-1')
df
```

```python
df.drop('laptop_ID' ,axis=1 , inplace=True )
```

```python
df.info()
```

```python
df.drop(['Product'] , axis=1 , inplace=True)
cols = df.columns[:-2]
for col in cols:
    print(f' {col} -> {df[col].unique()} \n\n')
```

```python
df = pd.concat([df , (pd.get_dummies(df['Company']))] , axis=1)
df = pd.concat([df , (pd.get_dummies(df['TypeName']))] , axis=1)
df.drop(['Company' , 'TypeName'] , axis=1 , inplace=True)
df.head()
```

```python
df['ScreenResolution'] = df['ScreenResolution'].str.split().apply(lambda x:x[-1])
df['Screen_W'] = df['ScreenResolution'].str.split('x').apply(lambda x:x[0]).astype('int')
df['Screen_h'] = df['ScreenResolution'].str.split('x').apply(lambda x:x[1]).astype('int')
df.drop(['ScreenResolution'] , axis=1 , inplace=True)
```

```python
df['Cpu_brand'] = df['Cpu'].str.split(' ').apply(lambda x:x[0])
df['Cpu_freq'] = df['Cpu'].str.split(' ').apply(lambda x:x[-1][:-3]).astype('float32')
df.drop(['Cpu'] , axis=1 , inplace=True)
```

```python
df['Ram'] = df['Ram'].str[:-2].astype('int32')
```

```python
df['Memory_amount'] = df['Memory'].str.split().apply(lambda x:x[0])
df['Memory_type'] = df['Memory'].str.split().apply(lambda x:x[1])
df.drop(['Memory'] , axis=1 , inplace=True)

def func(x):
    if 'GB' in x:
        return float(x[:x.find('GB')])
    elif 'TB' in x:
        return float(x[:x.find('TB')])*1000

df['Memory_amount'] = df['Memory_amount'].apply(func) 
df = pd.concat([df , pd.get_dummies(df['Memory_type'])] , axis=1)
df.drop(['Memory_type'] , axis=1 , inplace=True)
```

```python
df['Weight'] = df['Weight'].str[:-2].astype('float32')
```

```python
df['Gpu_brand'] = df['Gpu'].str.split().apply(lambda x: x[0])
df.drop(['Gpu'] , axis=1 , inplace=True)
temp = pd.get_dummies(df['Gpu_brand'])
temp.columns = [col +'_gpu' for col in temp.columns]
df = pd.concat([df , temp] , axis=1)
df.drop(['Gpu_brand'] , axis=1 , inplace=True)
```

```python
df = pd.concat([df , pd.get_dummies(df['OpSys'])] , axis=1)
df.drop(['OpSys'] , axis=1 , inplace=True)
```

```python
temp = pd.get_dummies(df['Cpu_brand'])
temp.columns = [col +'_cpu' for col in temp.columns]

df = pd.concat([df , temp] , axis=1)
df.drop(['Cpu_brand'] , axis=1 , inplace=True)
```

```python
df.head()
```

```python
features = list(df.corr()['Price_euros'].apply(abs).sort_values()[-30:].index)
```

```python
df[features].corr()['Price_euros']
```

```python
df.head()
```

```python
X , y = df[features].drop(['Price_euros'] , axis=1) , df['Price_euros']
X_train , X_test,y_train , y_test = tts(X , y , test_size=0.15)
sc = StandardScaler()
X_train = pd.DataFrame(sc.fit_transform(X_train))
X_test = pd.DataFrame(sc.transform(X_test))
rf = RandomForestRegressor(n_jobs=-1)
rf.fit(X_train , y_train)
rf.score(X_test , y_test)
```

```python
feature_cols = X.columns
feature_cols
```

```python
y_pred = rf.predict(X_test)
plt.scatter(y_test , y_pred)
```
